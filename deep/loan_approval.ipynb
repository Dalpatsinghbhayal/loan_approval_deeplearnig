```python
from tensorflow.keras.layers import Dense    #Dense is used to create a hidden layer
```

```python
from tensorflow.keras.models import Sequential
```

```python
x_train.shape
```

## Forward propagation

```python
ann_model = Sequential()
ann_model.add(Dense(units= 68, activation = 'relu',input_dim=11))   
                        #unit = 68 neurons honge
                        #we are adding a hidden layer using dense and defining 68 units chaiye
                        #pr isi mein input_dim = 11  x_train.shape ke i/p pe no col hote hain
                        #so i/p dim = 11
ann_model.add(Dense(units= 32, activation = 'relu'))
ann_model.add(Dense(units= 12, activation = 'relu'))
ann_model.add(Dense(units= 32, activation = 'relu'))

ann_model.add(Dense(units=1,))       #o/p layer bhi dense se hi define hoga but
                                     #binary hoga unti isiliye unit=1, 
                                     #activation function=sigmoid due to binary function
                                    #sigmoid give ouput in bw 0-1

```

## Backward propagation

```python
ann_model.compile()        #to add backward propagation
```
